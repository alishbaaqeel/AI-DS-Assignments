Assignment 3 – Exploratory Data Analysis (EDA)

Project Name: Credit Card Fraud Detection

Objective:
Perform data visualization on the credit card dataset to understand patterns, relationships, and distributions of features.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("cleaned_creditcard.csv")

print("Rows and Columns:", data.shape)
data.head()
```

```python
sns.set(style="whitegrid", palette="Set2")
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount ($)')
plt.ylabel('Frequency')
plt.show()
```

Most transactions are of small amounts, while very few are large — this is common in real-world credit card use.

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=data)
plt.title('Fraud vs Non-Fraud Transaction Count')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

The dataset is highly imbalanced — fraudulent transactions are extremely rare compared to normal ones.

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Class', y='Amount', data=data, estimator='mean')
plt.title('Average Transaction Amount by Class')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Average Amount ($)')
plt.show()
```

Fraud transactions usually have higher average amounts, suggesting that fraudsters target bigger payments.

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Features')
plt.show()
```

Some features (like V14, V17, and V12) show stronger correlation with the fraud class, which can help in model building later

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='V14', y='V17', hue='Class', data=data, alpha=0.5)
plt.title('Scatter Plot of V14 vs V17')
plt.xlabel('V14')
plt.ylabel('V17')
plt.legend(title='Class')
plt.show()
```

Fraudulent transactions often cluster separately from normal ones in certain feature combinations — useful for classification.
