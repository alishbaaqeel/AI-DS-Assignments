# Assignment 2 — Data Cleaning
## Project: Credit Card Fraud Detection
In this assignment, we cleaned the dataset by removing duplicates,
handling missing values, and treating outliers.


```python
import pandas as pd

# Load dataset
data = pd.read_csv("creditcard.csv")

# Show dataset info before cleaning
print("✅ BEFORE CLEANING:")
print("Rows and Columns:", data.shape)

# Show first 10 rows before cleaning
display(data.head(10))

```

Before Cleaning vs After Cleaning

This notebook shows the dataset before and after data cleaning.
We removed duplicates, handled missing values, and treated outliers.

```python
import numpy as np

# Count duplicate rows
duplicates = data.duplicated().sum()
print("Duplicate rows:", duplicates)

# Remove duplicates
data = data.drop_duplicates()
print("After removing duplicates:", data.shape)

# Check missing values
print("\nMissing values per column:")
print(data.isnull().sum())

# Fill missing values (if any)
data = data.fillna(data.mean(numeric_only=True))
print("\nAfter filling missing values:")
print(data.isnull().sum())

# Remove outliers using IQR
for col in data.select_dtypes(include=np.number).columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    data = data[(data[col] >= lower) & (data[col] <= upper)]

print("\nAfter removing outliers:", data.shape)

# Show first 10 rows after cleaning
print("\n✅ AFTER CLEANING:")
display(data.head(10))

# Save cleaned dataset
data.to_csv("cleaned_creditcard.csv", index=False)
print("\nCleaned dataset saved as cleaned_creditcard.csv ✅")
```
